#### Assignment: Simple Calculator

You are required to implement a simple calculator program in Python. The calculator should be able to perform basic arithmetic operations, such as addition, subtraction, multiplication, and division. Additionally, it should provide the option to exit the program.

Here are the steps to follow:

Create a new Python file called "calculator.py".

Define a function called "addition" that takes two numbers as input and returns their sum.

Define a function called "subtraction" that takes two numbers as input and returns their difference.

Define a function called "multiplication" that takes two numbers as input and returns their product.

Define a function called "division" that takes two numbers as input and returns their quotient. Remember to handle the case when the second number is zero (division by zero).

In the main part of the program, create a loop that allows the user to perform calculations until they choose to exit.

Inside the loop, display a menu with the following options:
a. Addition
b. Subtraction
c. Multiplication
d. Division
e. Exit

Prompt the user to enter their choice (a, b, c, d, or e) and read the input.

Depending on the user's choice, call the appropriate function (addition, subtraction, multiplication, or division) and display the result.

If the user chooses "Exit" (e), exit the loop and end the program.

Test your program by running it and performing various calculations.

```python
def addition(num1, num2):
    return num1 + num2    

def difference(num1,num2):
    return num1 - num2

def multiplication(num1, num2):
    return num1 * num2    

def division(num1, num2):
    if(num2 != 0):
        return num1/num2
    else:
        return 
```

```python
while 1:
    print("Options:")
    print("a. Addition")
    print("b. Subtraction")
    print("c. Multiplication")
    print("d. Division")
    print("e. Exit")
    
    operation = input("Enter your choice (a, b, c, d, or e):")
    try:
        num1 = float(input("Enter 1st number: "))
        num2 = float(input("Enter 2nd number: "))
    except ValueError:
        print("Enter valid numbers")
    
    if(operation == 'a'):
        print("Sum of the numbers is ",addition(num1, num2))
    elif(operation == 'b'):
        print("Difference of the numbers is ",difference(num1, num2))
    elif(operation == 'c'):
        print("Product of the numbers is ",multiplication(num1, num2))
    elif(operation == 'd'):
        print("Division of the numbers is ",division(num1, num2))
    elif(operation == 'e'):
        break
    else:
        print("Invalid option")
        continue  
        
```
