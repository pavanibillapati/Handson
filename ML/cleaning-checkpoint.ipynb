```python
import numpy as np

def remove_outliers_iqrmethod(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = (data > upper_bound) | (data < lower_bound)  # Use parentheses to group conditions
    
    cleaned_data = data[~outliers]  # Remove outliers based on the boolean mask
    
    return cleaned_data
```

```python
import numpy as np

def remove_outliers_IQR(data):
    
    Q1 = np.percentile(data,25)
    Q3 = np.percentile(data,75)
    
    IQR = Q3- Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 - 1.5 * IQR
    
    outliers = (data > upper_bound) | (data < lower_bound)
    outliers_count = ((data > upper_bound) | (data < lower_bound)).count()
    
    cleaned_df = data[~outliers]
    
    return cleaned_df, outliers_count
    
```
